## Xavier/Glorot Weight Initialization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../../datasets/ushape.csv', names=['X', 'Y', 'class'])
df.head()
```

```python
plt.scatter(df['X'], df['Y'], c=df['class'])
```

```python
X = df.iloc[:, 0:2].values
y = df.iloc[:, -1].values
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

## Glorot Uniform

```python
model1 = Sequential()

model1.add(Dense(5, activation='tanh', input_dim=2,
          kernel_initializer='glorot_uniform'))
model1.add(Dense(5, activation='tanh', kernel_initializer='glorot_uniform'))
model1.add(Dense(5, activation='tanh', kernel_initializer='glorot_uniform'))
model1.add(Dense(5, activation='tanh', kernel_initializer='glorot_uniform'))
model1.add(Dense(1, activation='sigmoid'))

model1.summary()
```

```python
model1.get_weights()
```

```python
model1.compile(loss='binary_crossentropy',
              optimizer='adam', metrics=['accuracy'])
```

```python
history = model1.fit(X, y, epochs=100, validation_split=0.2)
```

```python
model1.get_weights()
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X, y.astype('int'), clf=model1, legend=2)
```

## Glorot Normal

```python
model2 = Sequential()

model2.add(Dense(5, activation='tanh', input_dim=2,
          kernel_initializer='glorot_normal'))
model2.add(Dense(5, activation='tanh', kernel_initializer='glorot_normal'))
model2.add(Dense(5, activation='tanh', kernel_initializer='glorot_normal'))
model2.add(Dense(5, activation='tanh', kernel_initializer='glorot_normal'))
model2.add(Dense(1, activation='sigmoid'))

model2.summary()
```

```python
model2.get_weights()
```

```python
model2.compile(loss='binary_crossentropy',
              optimizer='adam', metrics=['accuracy'])
```

```python
history = model2.fit(X, y, epochs=100, validation_split=0.2)
```

```python
model2.get_weights()
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X, y.astype('int'), clf=model2, legend=2)
```
